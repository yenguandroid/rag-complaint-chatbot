# CrediTrust Financial — Complaint Intelligence Chatbot
## 1. Task 4 Overview
Task 4 delivers the user-facing chat interface for the CrediTrust Financial RAG complaint chatbot, making the system accessible to non-technical stakeholders — including Product Managers, Customer Support teams, and Compliance officers — without requiring any coding or data analysis skills. The interface was built using Gradio 6.19.0, deployed with a public shareable URL, and integrates seamlessly with the full RAG pipeline built across Tasks 1–3.

The completed application (app.py) provides a professional two-column chat experience: the left column displays the conversation history with the AI assistant, while the right column shows the retrieved complaint source cards that ground every answer in real CFPB complaint data.







## 2. Framework Selection — Comparative Analysis
### 2.1 Framework Comparison
Three candidate frameworks were evaluated for building the chat interface. Gradio was selected as the optimal choice for this use case based on its native chatbot component, minimal setup overhead, and built-in public sharing capability.



### 2.2 Why Gradio Was Chosen
-	Native gr.Chatbot component: Gradio provides a production-quality chat interface in a single component, eliminating the need to manually build message bubbles, conversation history management, or avatar rendering.
-	One-line public sharing: the share=True parameter in demo.launch() generates a public gradio.live URL instantly, making the demo shareable with stakeholders without any server configuration or deployment.
-	gr.HTML for source cards: the native HTML component allows rich, styled source display cards with complaint metadata — a key requirement for user trust and answer verification.
-	Minimal code footprint: the entire UI is implemented in under 150 lines of Python, making the codebase maintainable and easy to extend.
-	HuggingFace Spaces compatibility: Gradio apps deploy natively to HuggingFace Spaces for permanent hosting with zero additional configuration.
## 3. LLM Selection — Comparative Analysis
### 3.1 Models Evaluated
Four LLM options were evaluated during the development of Tasks 3 and 4. The selection was constrained by the available hardware: a Windows laptop with 8 GB RAM and no GPU. Local inference models proved impractical, leading to the adoption of the Groq API as the production LLM provider.



### 3.2 Why Groq API + Llama3 Was Chosen
-	Zero local RAM overhead: the model runs entirely in Groq's cloud infrastructure, consuming approximately 0 MB of local RAM — critical for the 8 GB machine constraint.
-	Sub-2-second responses: Groq's LPU (Language Processing Unit) hardware delivers responses in under 2 seconds per query, compared to 60–120 seconds for local CPU inference.
-	Free tier sufficient: the Groq free tier provides more than enough API calls for development, testing, and demonstration purposes at no cost.
-	Llama3-8b quality: the llama-3.1-8b-instant model delivers excellent instruction-following and context grounding — achieving an average quality score of 3.4/5 in the Task 3 evaluation, with correct honest refusals on aggregation queries.
-	Production upgrade path: for production deployment, the same code works with Groq's paid tier or can be switched to a different provider (OpenAI, Anthropic) by changing one line in rag_pipeline.py.
